```python
def plot_cluster_proportions(cluster_props,
                             cluster_palette=None,
                             xlabel_rotation=90,figdir=sc.settings.figdir,filename='proportion.png'):
    fig, ax = plt.subplots(dpi=300)
    fig.patch.set_facecolor('white')
    cmap = None
    if cluster_palette is not None:
        cmap = sns.palettes.blend_palette(
            cluster_palette,
            n_colors=len(cluster_palette),
            as_cmap=True)
    cluster_props.plot(
        kind='bar',
        stacked=True,
        ax=ax,
        legend=None,
        colormap=cmap
    )
    ax.legend(bbox_to_anchor=(1.01, 1), frameon=False, title='Type')
    sns.despine(fig, ax)
    ax.tick_params(axis='x', rotation=xlabel_rotation)
    ax.set_xlabel(cluster_props.index.name.capitalize())
    ax.set_ylabel('Proportion')
    fig.tight_layout()
    filepath = os.path.join(figdir,filename)
    print(filepath)
    plt.savefig(filepath)
    return fig
df = adata.obs[['leiden','concentration']].value_counts().unstack(level=1).fillna(0).apply(lambda x: x/x.sum()).apply(lambda x: x/x.sum(),axis=1)

plot_cluster_proportions(df,filename='barplot_leid.pdf')
plt.clf()

for i in df.index:
    dfi = df.loc[i]
    dfi.index = dfi.index.astype(str)+' ['+(dfi*100).apply(lambda x: np.round(x,1)).astype(str).apply(lambda x: x+'%')+']'
    fig = plt.figure(figsize=[5,5])
    ax = plt.subplot(1,1,1)
    dfi.plot.pie(subplots=True,figsize=[50,10],startangle=45,ax=ax, colors=adata.uns['concentration_colors'])
    fig.tight_layout()
    fig.savefig(f'figures/pie_{i}.pdf')
    del fig; plt.clf()
```
